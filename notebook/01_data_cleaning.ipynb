## DEP_DELAY vs DEP_DELAY_NEW

DEP_DELAY shows the exact difference between scheduled and actual departure time. 
Negative values mean early departure.

DEP_DELAY_NEW treats early departures as 0, measuring only actual delays.

For this analysis we will use DEP_DELAY_NEW, for calculating average delays and on-time rates, and DEP_DELAY for understanding full delay distribution.

## ARR_DELAY vs ARR_DELAY_NEW 
ARR_DELAY shows the exact difference between the actual and the schedule arrival time. Negative Value means earlt arrival.

ARR_DELAY_NEW treats early arrivals as 0, measuring only actual arrivals.

For this analysis we will use ARR_DELAY_NEW, for calculating average arrivals and on time rates, and ARR_DELAY for understanding full arrival distribuation.

## Feature: DepartureHour

Created DepartureHour by extracting hour 
from DEP_TIME column (HHMM format ÷ 100).

Cleaning decision:
87 flights had DEP_TIME = 2400 (midnight).
Converted hour 24 → hour 0 to maintain
valid 0-23 hour range.

Purpose:
Used for time of day delay pattern analysis
to identify which departure hours have
worst on-time performance.

```python
condition = [
    active_df['DEP_DELAY_NEW'] < 15,
    active_df['DEP_DELAY_NEW'].between(15, 60),
    active_df['DEP_DELAY_NEW'].between(61, 120),
    active_df['DEP_DELAY_NEW'] > 120
    
]

choices = [
    'On Time',
    'Minor',
    'Moderate',
    'Severe'
]

active_df['delaycategory'] = np.select(condition, choices, default='Unknown')
```

## Summary — Data Cleaning & Feature Engineering

### Cleaning Actions Performed

#### Data Type Fixes
- FL_DATE: object → datetime64 (enables date calculations)
- CANCELLED: float64 → int64 (binary flag, no decimals needed)
- DEP_DEL15: float64 → int64, renamed to IsDelayed
- ARR_DEL15: float64 → int64, renamed to IsArrDelayed

#### Rows Removed
- 2,902 diverted flights removed (ARR_DEL15 null)
  These flights departed but never arrived at destination
- Final active dataset: 1,082,277 flights

#### Outlier Investigation
- 24 flights with delays exceeding 24 hours investigated
- All 24 had valid recorded delay causes
- Most driven by LATE_AIRCRAFT_DELAY (cascade effect)
- Decision: Keep all 24 as valid operational events
- Maximum delay: 2,672 minutes (44 hours) — Envoy Air DTW→LGA

#### Column Name Fixes
- Stripped whitespace from all column names
- Renamed depaturehour → DepartureHour
- Renamed delaycategory → DelayCategory
- Consistent CamelCase for all engineered features

#### NULL Value Treatment
- Delay cause columns (CARRIER_DELAY, WEATHER_DELAY,
  NAS_DELAY, SECURITY_DELAY, LATE_AIRCRAFT_DELAY)
  filled with 0 for on-time flights
  Reasoning: 0 means no delay, not unknown data

#### 1. IsDelayed
- Definition: 1 if DEP_DELAY_NEW >= 15 minutes, else 0
- Industry standard FAA/BTS definition of delayed flight
- Result: 208,967 delayed (19.30%) | 873,310 on time (80.70%)

- Extracted from DEP_TIME (HHMM format ÷ 100)

  Severe:   > 120 mins →   30,097 flights  (2.78%)
- Business insight: 30,097 severe delays affect
